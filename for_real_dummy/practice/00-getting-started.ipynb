## 完整范例:验证 01 教程里的 softmax

回顾 [01-numpy-for-c-programmers.md](../01-numpy-for-c-programmers.md) 里的 softmax 实现。
验证不是跑一下看看输出顺不顺眼,而是写出你确定成立的性质,用 assert 检查。

softmax 有两个必然成立的性质:
1. 输出总和一定是 1(因为它是归一化后的概率分布)
2. 输入最大的那个位置,输出也一定最大(softmax 不改变大小顺序)

下面的代码把这两条都写成了 assert。

```python
def softmax(x):
    e = np.exp(x - x.max())
    return e / e.sum()

x = np.array([2.0, 1.0, 0.5, 3.0])
out = softmax(x)
print('softmax(x) =', out)

assert np.allclose(out.sum(), 1.0), f'总和是{out.sum()},应该是1'
assert out.argmax() == x.argmax(), 'softmax不应该改变最大值的位置'

print('两条性质都验证通过')
```
