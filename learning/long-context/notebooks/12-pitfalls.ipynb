## 核心陷阱

- Lost in middle: NIAH U 曲线 (Liu 2023)
- KV cache 1B model + 128k ≈ 17 GB
- 广告 ctx ≠ 可用 ctx
- 长 ctx + RAG 配合使用

```python
# 估算 KV cache 显存
def kv_cache_size_gb(n_layer, n_kv_head, head_dim, ctx_len, dtype_bytes=2):
    total_bytes = n_layer * n_kv_head * head_dim * 2 * ctx_len * dtype_bytes
    return total_bytes / 1e9
for L in [4096, 32768, 131072]:
    print(f'  ctx={L}: KV={kv_cache_size_gb(32, 8, 128, L):.1f} GB')
```
