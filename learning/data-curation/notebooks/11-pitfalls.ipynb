## 陷阱 1: benchmark contamination 演示

```python
benchmark = 'Janet had 16 eggs. She gave 5 to her sister.'
corpus = ['random web doc', benchmark + ' She sold 3 at the market.', 'another doc']
# 13-gram exact match check
def ngrams(text, n=13):
    words = text.split()
    return set(' '.join(words[i:i+n]) for i in range(len(words)-n+1))

bench_grams = ngrams(benchmark)
for d in corpus:
    contaminated = any(g in d for g in bench_grams)
    print('CONTAMINATED' if contaminated else 'clean', '<-', d[:50])
```
