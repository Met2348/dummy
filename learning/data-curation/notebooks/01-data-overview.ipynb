## 主流语料时间线可视化

```python
import matplotlib.pyplot as plt
rows = [
    ('C4',          2019,  750e9 / 4),     # ~ tokens
    ('Pile',        2020,  825e9 / 4),
    ('RedPajama',   2023,  1.2e12),
    ('SlimPajama',  2023,  627e9),
    ('FineWeb',     2024,  15e12),
    ('FineWeb-Edu', 2024,  1.3e12),
    ('DCLM',        2024,  4e12),
    ('Dolma',       2024,  3e12),
]
fig, ax = plt.subplots(figsize=(9,4))
for name, year, tok in rows:
    ax.scatter(year, tok / 1e9)
    ax.annotate(name, (year, tok/1e9), fontsize=8)
ax.set_yscale('log')
ax.set_ylabel('tokens (B)')
ax.set_xlabel('year')
ax.set_title('Public LLM corpus scale over time')
plt.show()
```
