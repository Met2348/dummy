```python
import sys, tempfile, os; sys.path.insert(0,'../src')
from dataset_shards import write_shard, ShardManager
with tempfile.TemporaryDirectory() as d:
    for i in range(3):
        ids = list(range(i*1000, (i+1)*1000))
        write_shard(ids, os.path.join(d, f'shard_{i}.bin'))
    paths = [os.path.join(d, f'shard_{i}.bin') for i in range(3)]
    mgr = ShardManager(paths)
    for _ in range(3):
        seq = mgr.next_seq(100)
        print(f'state={mgr.state()} first={seq[0]}')
```
