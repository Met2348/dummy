```python
import sys, os
from pathlib import Path
# 把专题 src/ 加入路径, 以便 import 工具
SRC = Path.cwd().parent / "src" if Path.cwd().name == "notebooks" else Path.cwd() / "src"
sys.path.insert(0, str(SRC))
import make_cards, citation_graph as cg
print("工具就绪:", "make_cards", "citation_graph")

PAPER = "Direct Preference Optimization"   # ← 换成你要解剖的论文
```

## Pass 1 · 鸟瞰 (5C, 5–10 min)
只看 标题/摘要/引言 + 节标题 + 扫公式 + 结论 + 参考文献。回答 5C, 然后决定弃/留。

```python
pass1 = {
    "Category 类别":        "新方法 (alignment, 替代 RLHF 的偏好优化)",
    "Context 脉络":         "接在 RLHF(InstructGPT)/PPO 之后; 理论基础=Bradley-Terry 偏好模型",
    "Correctness 初判":     "把 RLHF 目标解析地重参数化为分类损失, 推导看起来站得住",
    "Contributions 自称":   "免 reward model、免 RL、单阶段, 却能匹配 PPO 效果",
    "Clarity 清晰度(1-5)":  4,
    "决策":                 "进 Pass2 (与我方向核心相关)",
}
for k, v in pass1.items():
    print(f"{k:18s}: {v}")
```

## Pass 2 · 抓骨架 (~1h)
懂"它做了什么、证据是什么"。合上论文能向同门复述, 才算过关。

```python
pass2 = {
    "一句话问题": "RLHF 又重又不稳(要训 reward model + 跑 PPO), 能不能更简单?",
    "一句话方法": "用偏好对直接优化策略, 损失等价于一个带参考模型的二分类",
    "关键证据":   "在情感/摘要/对话上匹配或超过 PPO, 且训练更稳更省",
    "能否复述":   "能",
    "需补的背景": ["Bradley-Terry 模型", "PPO/KL 约束", "RLHF 三段式"],
}
for k, v in pass2.items():
    print(f"{k}: {v}")
```

## 引用邻域 · 真实科研动作
拉这篇的 **references**(它站在谁肩上) 和 **citations**(谁在它之上继续做)。
无网时自动回退到内置 DPO 样例, 保证出图。看右侧"引它的"那列 —— 那是找 gap 的金矿。

```python
# 用 kernel 默认 inline 后端: 图直接嵌进 notebook (headless 也可), 无散落文件
import matplotlib.pyplot as plt

data = cg.fetch_paper_graph(PAPER, limit=8)
print(cg.summarize(data))

G = cg.build_graph(data)
ax = cg.plot_graph(G)
plt.tight_layout()
print("\n节点数:", G.number_of_nodes(), "| 边数:", G.number_of_edges())
plt.show()
```

## Pass 3 + 10 问攻击清单
切换成"想拒掉这篇"的审稿人。逐条命中即记 —— 每个命中点都是潜在 gap。

```python
import pandas as pd
attack = [
    ("Q1 baseline 公平?",      False, "对比了 PPO, 算较强基线"),
    ("Q2 算力/数据偷加?",      False, "同数据下对比, 基本公平"),
    ("Q3 统计显著性?",         True,  "部分结果种子数偏少, 方差报告不全 ← 可攻击"),
    ("Q4 测试集干净?",         False, "任务设定下泄露风险低"),
    ("Q5 指标偏心?",           False, "用了任务标准指标"),
    ("Q6 覆盖够?",             True,  "规模/任务有限, 大模型上是否成立未充分验证 ← 可攻击"),
    ("Q7 消融做全?",           True,  "对 β、参考模型选择的消融可更系统 ← 可攻击"),
    ("Q8 结论没超证据?",       False, "宣称克制"),
    ("Q9 可复现?",             False, "方法简单、社区复现多"),
    ("Q10 它没测什么?",        True,  "偏好数据有噪声时如何? 几乎没测 ← 最大机会"),
]
df = pd.DataFrame(attack, columns=["问题", "命中(可攻击)", "备注"])
print(df.to_string(index=False))
hits = df[df["命中(可攻击)"]]
print(f"\n共 {len(hits)} 个可攻击点 → 下一步可开 gap 卡 (见 N2)")
```
