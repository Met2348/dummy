```python
import sys, os
sys.path.insert(0, os.path.abspath('../src'))
from harness.context import ContextWindow
w = ContextWindow(budget=40)
w.add('system', 'you are an agent')
for i in range(10): w.add('user', f'msg {i} with padding content here')
print('before:', w.used())
reclaimed = w.compact(keep_recent=2)
print('reclaimed:', reclaimed, '-> after:', w.used())
print([m['role'] for m in w.messages])
```
