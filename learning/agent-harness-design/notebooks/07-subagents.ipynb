```python
import sys, os
sys.path.insert(0, os.path.abspath('../src'))
from harness.subagents import run_subagent, fan_out
from harness.model import MockModel, ModelResponse, ToolCall
from harness.tools import ToolRegistry
reg = ToolRegistry().add('lookup', 'look up', lambda topic: f'fact:{topic}', read_only=True)
def make(goal):
    def brain(msgs):
        res = {m['name']: m['content'] for m in msgs if m.get('role')=='tool'}
        if 'lookup' not in res: return ModelResponse('q', [ToolCall('lookup', {'topic': goal})])
        return ModelResponse('sum: ' + res['lookup']['value'])
    return brain
print(fan_out(['a','b'], lambda g: MockModel(make(g)), reg))
```
