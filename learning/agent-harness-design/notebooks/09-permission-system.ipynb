# Permission modes

Topic 9 · agent-harness-design. 运行下方单元。

```python
import sys, os
sys.path.insert(0, os.path.abspath('../src'))
from harness.permissions import PermissionManager
from harness.tools import Tool
write = Tool('write_file', 'writes', lambda: None)
read = Tool('read_file', 'reads', lambda: None, read_only=True)
for mode in ('auto', 'readonly'):
    pm = PermissionManager(mode=mode)
    print(mode, '-> write:', pm.check(write).action, '| read:', pm.check(read).action)
```
