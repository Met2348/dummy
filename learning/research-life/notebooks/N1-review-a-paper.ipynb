# N1 · 给一篇论文写结构化评审 + 调转枪口审自己 (Review a Paper)

> 配套 9.9-L1 · **真实科研动作**: 用一份固定 rubric 给一篇工作写**建设性、结构化**的评审,
> 跑建设性自检, 再把同一套维度**调转枪口审视自己的工作** —— 审稿对写作的最直接反哺。

材料用你熟悉的 Robust-DPO 工作 (也可换成任何你在审/在写的论文)。

```python
import sys
from pathlib import Path
SRC = Path.cwd().parent / "src"
sys.path.insert(0, str(SRC))
import review_kit as rv
print("评审维度:", [rv.DIMENSIONS[k][0] for k in rv.DIMENSIONS])
```

## 1. 写一份结构化评审 (先复述, 再逐维度: 分+证据+建议)

注意建设性三件套: summary 复述 / 负面带证据 / 负面带可操作建议。

```python
r = rv.blank_review()
r["summary"] = "提出 Robust-DPO: 在偏好标签有噪声时对可疑样本降权, 提升对齐 win-rate。"
r["dimensions"]["soundness"] = {"score": 4,
    "evidence": "8 种子报方差, baseline 公平调参, 交互效应 p<0.001",
    "suggestion": "建议补不同噪声类型 (现仅一种), 增强泛化论证"}
r["dimensions"]["novelty"] = {"score": 3,
    "evidence": "鲁棒加权思路与已有 robust-DPO 工作相近",
    "suggestion": "在相关工作里更清楚地和最像的那篇划界 (9.2 领域地图)"}
r["dimensions"]["clarity"] = {"score": 4, "evidence": "图清晰, 交互效应一目了然", "suggestion": ""}
r["dimensions"]["reproducibility"] = {"score": 5, "evidence": "代码/超参/种子齐全", "suggestion": ""}
r["dimensions"]["significance"] = {"score": 4,
    "evidence": "噪声偏好是真实痛点", "suggestion": ""}
r["overall_score"] = 4; r["confidence"] = 4
print(rv.render(r))
```

## 2. 建设性自检 (低分必须带证据 + 可操作建议, 否则不合格)

```python
chk = rv.constructiveness_check(r)
if chk["constructive"]:
    print("✅ 这是一份建设性评审 (每个判断都有证据 + 改进方向)")
else:
    for i in chk["issues"]:
        print("⚠", i)
print("\n对比: '方法不新颖, 拒。' 这种没证据没建议的评审 = 懒惰且无用 (9.9-L1)。")
```

## 3. 反面: 一份糟糕的评审 (给低分却不说理由)

```python
bad = rv.blank_review()  # 连 summary 都没写
bad["dimensions"]["novelty"] = {"score": 1, "evidence": "", "suggestion": ""}
for i in rv.constructiveness_check(bad)["issues"]:
    print("⚠", i)
print("→ 缺复述 + 低分无证据无建议 = 你 9.7 时恨过的那种烂审稿。别成为它。")
```

## 4. 调转枪口: 用同样维度审自己 (审稿对写作的反哺)

审完别人, 趁热用同一套苛刻视角审自己最近的工作。

```python
print("用评审维度审视我自己的工作:")
for q in rv.flip_to_self():
    print("  ?", q)
print("\n→ 你刚以审稿人视角校准好了眼力, 趁热审自己能抓出平时看不见的破绽。")
print("→ 这连回 9.3 攻击清单 / 9.7 narrative_audit: 同一套批判视角的闭环。")
```
