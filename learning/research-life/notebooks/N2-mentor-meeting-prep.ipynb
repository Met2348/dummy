## 1. 一份准备充分的议程 (进展/卡点/需决策/下一步)

关键: ② 卡点要**具体**, ③ 决策要**带选项 + 你的倾向** (别开放式"咋办")。

```python
agenda = mp.build_agenda(
    progress=[
        "跑完 Robust-DPO 噪声消融: 40% 噪声下 +13 点 (p<0.001, 8 种子)",
        "复现 baseline DPO 并同等调参 (回应公平性, 9.4-L3)",
    ],
    blockers=[
        "想加 70B 实验但显存不够, 试了梯度检查点和 LoRA 都还 OOM —— 该放弃 70B 还是申请更多卡?",
    ],
    decisions=[
        "投 EMNLP (6月ddl, 时间紧) 还是 NeurIPS (5月ddl, 更对口)? 我倾向 EMNLP, 因为能赶上且 Findings 兜底",
    ],
    next_steps=["补不同噪声类型消融", "写 method 节初稿 (用 9.7 装配)"],
)
print(mp.render(agenda))
```

## 2. 自检: 有没有明确的 ask?

```python
rep = mp.audit(agenda)
if rep["ready"]:
    print("✅ 准备就绪: 有明确 ask (卡点 + 带选项的决策), 不是纯汇报")
else:
    for i in rep["issues"]:
        print("⚠", i)
```

## 3. 反面教材: 纯汇报型 meeting (浪费双方时间)

新手最常见: 去了就流水账汇报, 没有任何要导师帮忙决定/解锁的东西。

```python
bad = mp.build_agenda(progress=["这周看了几篇论文", "跑了一些实验"])
for i in mp.audit(bad)["issues"]:
    print("⚠", i)
print("\n→ 没有 ask 的 meeting 基本是浪费导师 30 分钟。")
print("→ 去之前先想清: 我要导师帮我【决定】什么 / 【解锁】什么 / 【确认】什么? (9.9-L2)")
```
