## 相对位置不变性验证

```python
torch.manual_seed(0)
d = 16
q = torch.randn(1, d); k = torch.randn(1, d)
cos, sin = rope.build_cos_sin(t=20, dim=d)
scores = []
for m, n in [(0,3), (5,8), (10,13)]:
    qr = rope.apply_rope_interleaved(q, cos[m:m+1], sin[m:m+1])
    kr = rope.apply_rope_interleaved(k, cos[n:n+1], sin[n:n+1])
    scores.append((qr * kr).sum().item())
print('scores at same relative dist=3:', scores)
print('diff:', max(scores) - min(scores))
```
