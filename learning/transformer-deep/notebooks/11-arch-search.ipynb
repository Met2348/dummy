# L11 · μP + architecture search

概念讲解 lecture。这里只做 architecture 参数计算 demo。

```python
# 启发式 d_model = 64 * sqrt(N / 7e6)
import math
for N_M in [100, 1000, 7000, 70000]:
    d = 64 * math.isqrt(int(math.sqrt(N_M * 1e6 / 7e6) * 64))
    d = round(d / 64) * 64
    h = d // 64
    n_layer = round(N_M * 1e6 / (12 * d * d))
    print(f'{N_M:>6}M params: d_model={d}  n_head={h}  n_layer={n_layer}')
```
