```python
# 复习公式：Attention(Q,K,V) = softmax(QK^T/sqrt(d)) V
import torch, math
d = 64
q = torch.randn(1, 4, d)
k = torch.randn(1, 4, d)
v = torch.randn(1, 4, d)
score = q @ k.transpose(-2, -1) / math.sqrt(d)
attn = score.softmax(dim=-1)
out = attn @ v
print('shapes:', score.shape, attn.shape, out.shape)
```
