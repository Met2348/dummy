## Sinusoidal PE 可视化

```python
pe = pe_sinusoidal.sinusoidal_pe(t=64, d=32)
plt.figure(figsize=(8,3))
plt.imshow(pe.numpy(), aspect='auto', cmap='RdBu')
plt.colorbar(); plt.title('Sinusoidal PE'); plt.xlabel('dim'); plt.ylabel('pos')
plt.show()
```

## ALiBi bias 可视化

```python
bias = pe_alibi.alibi_bias(t=16, num_heads=8)
fig, ax = plt.subplots(2, 4, figsize=(12, 5))
for h in range(8):
    ax[h//4, h%4].imshow(bias[h].numpy(), cmap='RdBu', vmin=-8, vmax=0)
    ax[h//4, h%4].set_title(f'head {h}')
plt.tight_layout(); plt.show()
```
