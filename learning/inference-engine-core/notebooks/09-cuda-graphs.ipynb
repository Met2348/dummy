本节为概念讲解 + 真实 CUDA Graphs API 示例（需 GPU 才可跑）。

```python
# 概念伪代码 (CPU 跑不动 cuda.graph)
import torch
print('cuda available:', torch.cuda.is_available())
if torch.cuda.is_available():
    static_in = torch.zeros(1, 16, device='cuda')
    static_out = torch.zeros(1, 16, device='cuda')
    # capture
    g = torch.cuda.CUDAGraph()
    with torch.cuda.graph(g):
        static_out.copy_(static_in * 2 + 1)
    # replay
    static_in.copy_(torch.ones(1,16,device='cuda'))
    g.replay()
    print('out:', static_out[0,:5])
```
