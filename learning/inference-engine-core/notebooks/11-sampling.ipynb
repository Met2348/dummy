```python
import sys; sys.path.insert(0,'../src')
import torch
from sampling import SamplerConfig, sample_one
torch.manual_seed(0)
logits = torch.randn(100)
for T in [0, 0.7, 1.5]:
    cfg = SamplerConfig(temperature=T, top_k=10, top_p=0.9)
    samples = [sample_one(logits, torch.tensor([], dtype=torch.long), cfg) for _ in range(20)]
    print(f'T={T}: {samples[:8]}...')
```
