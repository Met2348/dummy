# L10 · RL Pitfalls

对应 lecture: `lectures/10-rl-pitfalls.md`

演示几个典型 pitfall：
 - reward over-optimization 曲线（合成数据）
 - mode collapse signal
 - PPO 超参敏感性可视化（用之前的 ppo_tricks_ablation 数据）


```python
import numpy as np
import matplotlib.pyplot as plt

# 模拟 Gao 2022 over-optimization 曲线
steps = np.linspace(0, 100, 100)
proxy = 0.5 + 0.4 * (1 - np.exp(-steps/30))  # RM score 单调上升
gold  = 0.5 + 0.3 * (1 - np.exp(-steps/20)) - 0.2 * (steps/100)**2  # 真人偏好先升后降

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(steps, proxy, label='proxy (RM score)', linewidth=2)
ax.plot(steps, gold,  label='gold (human preference)', linewidth=2)
ax.set_xlabel('RL step'); ax.set_ylabel('reward')
ax.set_title('Reward Over-optimization (Gao 2022 风格)')
ax.legend(); ax.grid(alpha=0.3)
ax.axvline(x=steps[np.argmax(gold)], color='red', linestyle='--', alpha=0.5,
           label='best gold')
ax.legend()
plt.tight_layout(); plt.show()
```

## 模式塌缩示例（entropy curve）

```python
steps = np.linspace(0, 100, 100)
good_run = 0.69 * np.exp(-steps/200) + 0.05    # 缓慢降到 0.05
collapse_run = 0.69 * np.exp(-steps/15)        # 快速降到 ~0

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(steps, good_run, label='healthy: entropy → 0.05', linewidth=2)
ax.plot(steps, collapse_run, label='collapse: entropy → 0', linewidth=2, color='red')
ax.set_xlabel('iter'); ax.set_ylabel('mean entropy')
ax.set_title('Mode Collapse 信号')
ax.axhline(y=0.01, color='gray', linestyle=':', label='collapse threshold')
ax.legend(); ax.grid(alpha=0.3)
plt.tight_layout(); plt.show()
```
