# L06 · PPO 7 件套 ablation

对应 lecture: `lectures/06-ppo-tricks.md`

跑 5 配置（每个 30k step，约 10 分钟）观察 adv norm / grad clip / orthog init 的影响。


```python
import sys, os, copy
sys.path.insert(0, os.path.abspath('../src'))
from ppo_tricks_ablation import run_ppo

base = dict(
    env='CartPole-v1', n_envs=8, n_steps=128, hidden=64,
    lr=3e-4, gamma=0.99, lam=0.95, eps=0.2,
    vf_coef=0.5, ent_coef=0.01, max_grad_norm=0.5,
    K_epochs=10, n_minibatches=4,
    adv_norm=True, orthog_init=True,
    total_steps=30_000, seed=42,
)

configs = [
    ('baseline', {}),
    ('-adv_norm', dict(adv_norm=False)),
    ('-grad_clip', dict(max_grad_norm=None)),
    ('-orthog', dict(orthog_init=False)),
]
results = []
for name, override in configs:
    cfg = copy.deepcopy(base); cfg.update(override)
    r = run_ppo(cfg)
    results.append((name, r))
    print(f'{name:<14} → {r:.1f}')
```

```python
import matplotlib.pyplot as plt
names = [r[0] for r in results]
vals = [r[1] for r in results]
plt.figure(figsize=(8, 4))
plt.bar(names, vals)
plt.ylabel('Mean reward (last 100 ep)')
plt.title('PPO 7 件套消融 — CartPole 30k steps')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout(); plt.show()
```
