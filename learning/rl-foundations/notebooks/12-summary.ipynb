 - PPO 在所有 LLM-RL 中的角色


```python
import matplotlib.pyplot as plt
import numpy as np

# 12 方法在不同维度的雷达图（示意）
algos = ['REINFORCE', 'A2C', 'TRPO', 'PPO', 'PPO+KL']
dims = ['data eff', 'variance ↓', 'KL controll', 'LLM ready', 'simplicity']
scores = np.array([
    [1, 2, 1, 1, 5],   # REINFORCE
    [3, 3, 2, 3, 4],   # A2C
    [4, 4, 5, 2, 1],   # TRPO
    [5, 4, 4, 4, 4],   # PPO
    [5, 4, 5, 5, 3],   # PPO+KL
])
fig, ax = plt.subplots(figsize=(10, 4))
x = np.arange(len(dims))
width = 0.15
for i, algo in enumerate(algos):
    ax.bar(x + i*width, scores[i], width, label=algo)
ax.set_xticks(x + width*2)
ax.set_xticklabels(dims)
ax.set_ylabel('score (1-5)'); ax.legend(); ax.grid(axis='y', alpha=0.3)
ax.set_title('RL Foundations - 5 算法横向（5 维评分）')
plt.tight_layout(); plt.show()
```
