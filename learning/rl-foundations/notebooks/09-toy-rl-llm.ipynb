# L09 · Toy RL-LLM (sentiment reward)

对应 lecture: `lectures/09-toy-rl-llm.md`

 - 调用 sentiment_reward.SentimentReward 当 RM
 - 观察 BERT-sentiment 对不同样本的打分


```python
import sys, os
sys.path.insert(0, os.path.abspath('../src'))

from sentiment_reward import SentimentReward
rm = SentimentReward(device='cpu')

samples = [
    'The movie was absolutely amazing!',
    'good good good good good good',
    'The movie was terrible and boring.',
    'It was okay, average movie I guess.',
]
scores = rm.score(samples)
for txt, s in zip(samples, scores.tolist()):
    print(f'{s:.3f}   {txt}')
```

## 观察 reward hacking 趋势

注意第 2 行 `'good good good good good good'`：BERT 给的分数与第 1 句相近，这就是 reward hacking 的根源。

训 PPO 不加 KL penalty 时，GPT-2 会发现这类捷径。
