```python
import sys, os
sys.path.insert(0, os.path.abspath('../src'))
import numpy as np
import matplotlib.pyplot as plt
from gae import gae_numpy, gae_lambda_zero_equals_td, gae_lambda_one_equals_mc
```

```python
# Lecture 例子
rewards = np.array([1.0, 1.0, 1.0])
values = np.array([0.5, 0.3, 0.2])
dones = np.array([0.0, 0.0, 1.0])

A_0, _ = gae_numpy(rewards, values, dones, 0.0, 1.0, 0.0)
A_5, _ = gae_numpy(rewards, values, dones, 0.0, 1.0, 0.95)
A_1, _ = gae_numpy(rewards, values, dones, 0.0, 1.0, 1.0)
print('λ=0    A =', A_0)
print('λ=0.95 A =', A_5)
print('λ=1    A =', A_1)
print('expected (λ=0.95): [2.377, 1.66, 0.8]')
```

```python
# 长 trajectory 可视化
rng = np.random.default_rng(0)
T = 50
rewards = rng.standard_normal(T)
values = rng.standard_normal(T) * 0.5
dones = np.zeros(T); dones[-1] = 1

lams = [0.0, 0.5, 0.95, 1.0]
fig, ax = plt.subplots(figsize=(10, 4))
for lam in lams:
    A, _ = gae_numpy(rewards, values, dones, 0.0, 0.99, lam)
    ax.plot(A, label=f'λ={lam}')
ax.set_xlabel('t'); ax.set_ylabel('advantage'); ax.legend(); ax.grid(alpha=0.3)
ax.set_title('GAE advantage under different λ')
plt.show()
```
