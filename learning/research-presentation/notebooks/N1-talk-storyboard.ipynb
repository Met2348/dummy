# N1 · 排一个 talk 分镜 (Talk Storyboard)

> 配套 9.8-L1/L2 · **真实科研动作**: 给你的研究排一个有时间预算的 talk 分镜, 跑结构自检 ——
> 有没有唯一 takeaway? motivation 超不超长? slide 过不过密?

```python
import sys
from pathlib import Path
SRC = Path.cwd().parent / "src"
sys.path.insert(0, str(SRC))
import talk_planner as tp
print("标准 talk 结构:", [name for _,name,_,_ in tp.TALK_STRUCTURE])
```

## 1. 先定唯一 takeaway, 再排分镜 (L1 最高约束)

takeaway 是一个 talk 的灵魂: 听众离场后唯一记住的一句话。先写它。

```python
takeaway = "Robust-DPO 的优势随偏好标签噪声增大而扩大"
# 给一个 12 分钟 talk 排分镜, 并填上每部分打算用几张 slide
storyboard = tp.plan(12, takeaway,
                     slides={"motivation": 3, "approach": 4, "result": 5, "takeaway": 2})
print(tp.render(storyboard, tp.audit(storyboard)))
```

## 2. 反面教材: motivation 太长 + 忘了 takeaway

新手最常见的两个病, 看自检能不能抓出来。

```python
bad = tp.plan(10, "")                       # 忘了写 takeaway
bad["parts"][0]["minutes"] = 4.0           # 强行把 motivation 拉到 40%
report = tp.audit(bad)
print(tp.render(bad, report))
print("\n→ 自检抓出了: 无 takeaway + motivation 过长 (新手两大通病)。")
```

## 3. slide 过密自检 (一图一点, 别堆太多)

如果 12 分钟塞 25 张 slide, 每张停留 < 30 秒, 听众跟不上。看自检提示。

```python
dense = tp.plan(12, takeaway,
                slides={"motivation": 6, "approach": 8, "result": 8, "takeaway": 3})  # 25 张!
rep = tp.audit(dense)
for i in rep["issues"]:
    print("⚠", i)
print("→ 解药: 砍 slide 或加时间。一张 slide 一个点, 宁少勿密 (L2)。")
```

## 4. 落地成分镜模板 (存档进你的第二大脑, 9.1)

```python
out = Path.cwd() / "_talk_output"; out.mkdir(exist_ok=True)
(out / "storyboard.md").write_text(tp.render(storyboard, tp.audit(storyboard)), encoding="utf-8")
print("已存 storyboard.md; 接下来照它逐页做 slide (一页一个论断标题)。")
```

## 5. 反思

你刚把一个研究排成了卡死时间、结构自检通过的 talk 分镜。带走:
- **先定唯一 takeaway**, 它是 talk 的最高约束 (L1)。
- 时间预算: motivation 别超 30% (新手通病), takeaway 永远保住。
- slide 一图一点、宁少勿密。

> 真用时: 照分镜逐页做 slide (PPT/Keynote/beamer), 每页标题写成**论断**, 搬 9.6 的图,
> 然后**对着计时器排练 5 遍** —— 工具排不了的那部分 (临场), 只能靠练 (L2)。
