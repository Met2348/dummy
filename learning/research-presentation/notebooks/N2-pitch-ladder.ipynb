> 配套 9.8-L3 · **真实科研动作**: 把研究压成 10秒/1分钟/5分钟三档, 写「外行版」vs「专家版」,
> 用工具看时长估算 + 黑话检测。能把研究讲成一句话, 是「想清楚了」的证据。

```python
import sys
from pathlib import Path
SRC = Path.cwd().parent / "src"
sys.path.insert(0, str(SRC))
import pitch_kit as pk
print("三档字数预算 (中文 ~220 字/分):")
for lv in pk.LADDER:
    lo, hi = pk.char_budget(lv)
    print(f"  {lv:<5} {lo}-{hi} 字")
```

## 1. 10 秒钩子: 外行版 vs 专家版 (同一研究, 不同语言)

L3 核心: 按听众调术语密度。给外行满嘴黑话 = 没人懂。

```python
hook_layperson = "我教 AI 在人给的反馈不靠谱、有噪声时, 还能学好怎么回答问题。"
hook_expert    = "noise-robust DPO: 偏好标签有噪声时仍保住对齐 win-rate。"

print("【外行版】", hook_layperson)
print(pk.render_check(hook_layperson, "10s", audience="layperson"))
print("\n【专家版】", hook_expert)
print(pk.render_check(hook_expert, "10s", audience="peer"))
```

## 2. 反面教材: 给外行用专家版 (黑话翻车)

把专家版讲给外行 —— 工具会揪出听不懂的黑话。

```python
print("把专家版讲给外行会怎样:")
print(pk.render_check(hook_expert, "10s", audience="layperson"))
print("→ DPO/对齐/鲁棒 对外行是天书。换成大白话 (上面的外行版) 才有效。")
```

## 3. 1 分钟版 (问题→做了什么→为什么重要)

```python
one_min = (
    "现在主流的大模型对齐方法, 比如 DPO, 都默认人给的偏好标注是干净、一致的。"
    "但现实里这些标注充满噪声和分歧 —— 不同标注者意见相左、甚至标错。一有噪声, "
    "DPO 的效果就明显下滑。我做了一个改进版, 在训练时对可疑的偏好对降低权重, "
    "结果是: 噪声越大, 它相对 DPO 的优势越明显, 40% 噪声下能高出十几个点。"
    "这让对齐方法在真实的、不完美的数据上更可靠。"
)
print(one_min)
print(pk.render_check(one_min, "1min", audience="peer"))
```

## 4. 同心圆检查: 三档是否层层包含

```python
print("同心圆结构 (L3): 10s 内核 → 1min 加问题/意义 → 5min 加方法/证据")
print(f"  10s ({pk._char_count(hook_layperson)}字): 钩子")
print(f"  1min ({pk._char_count(one_min)}字): + 问题背景 + 量化结果 + 意义")
print("  5min: 再 + 方法直觉 + 那张交互效应图 + future work")
print("\n→ 对方兴趣到哪, 你讲到哪。先给最短的, 看反应再展开 (双向)。")
```
