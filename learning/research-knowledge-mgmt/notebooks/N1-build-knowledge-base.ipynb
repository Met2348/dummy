```python
import sys
from pathlib import Path
import pandas as pd

# 把本专题 src/ 加进路径 (nbconvert 执行时 cwd = 本 notebook 所在目录)
SRC = Path.cwd().parent / "src"
sys.path.insert(0, str(SRC))
import bib_to_cards
print("已载入 bib_to_cards, 源目录:", SRC)
```

## 1. 解析 BibTeX (这里用内置样例; 真用时换成你 Zotero 导出的 library.bib)

`bib_to_cards.SAMPLE_BIB` 内置了 3 篇你已复现过/熟悉的论文 (DPO / R1 / InstructGPT)。
真实使用时只需: `text = open("library.bib", encoding="utf-8").read()`。

```python
entries = bib_to_cards.parse_bibtex(bib_to_cards.SAMPLE_BIB)
df = pd.DataFrame([{
    "citekey": e.get("citekey"),
    "title": e.get("title", "")[:48] + ("…" if len(e.get("title",""))>48 else ""),
    "year": e.get("year"),
    "type": e.get("entrytype"),
} for e in entries])
print(f"解析出 {len(entries)} 条文献:")
df
```

## 2. 一键生成文献卡库

对每篇论文生成一张预填元数据、留好批注空位的 markdown 卡。
落到 `_kb_output/library/` (该目录已 gitignore, 是你的个人草稿区)。

```python
out = Path.cwd() / "_kb_output" / "library"
paths = bib_to_cards.make_cards(entries, out_dir=out)
print(f"生成 {len(paths)} 张文献卡 → {out.relative_to(Path.cwd())}/")
for p in paths:
    print("  -", p.name)
```
